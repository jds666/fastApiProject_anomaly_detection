## 数据分组

```python
## 温度数据
temperature_data = data[data['temperature'].notnull()==True]
## 温度的分组
Temp = temperature_data[{'point_no','ts','temperature'}]
Temp = Temp.dropna() ## 去除NAN值
Temp_point_no_groups = Temp.groupby(['point_no'])
print("一共包含  ",len(Temp_point_no_groups),"  Temp_point_no_groups")
```

```python
# 每一组存入csv
i=0
for point_no,point in Temp_point_no_groups:
    point = point.sort_values(by=['ts']).loc[:,['ts','temperature']]
    point.drop_duplicates(subset=['ts'], inplace=True)
    point.rename(columns={'ts':'timestamp','temperature':'value'}, inplace=True)
    point.to_csv('../data_csv/temperature/Temperature_point_no_'+str(point_no)+'.csv',index=False)
    print("保存point_no",point_no,"的数据到 Temperature_point_no_grouped_"+str(point_no)+".csv")
    print(i+1," : 总数据量 ",len(point)," 条")
    print("------------------------------------------------------------")
    print("")
    i=i+1
```

```python
## 有线数据
wired_data = data[(data['acceleration_peak_x'].notnull()&data['acceleration_peak_y'].isnull())==True]
## 有线数据分组
wired =wired_data[{'point_no','ts','acceleration_peak_x' ,'acceleration_rms_x','speed_peak_x','speed_rms_x'}]
wired = wired.dropna() ## 去除NAN值
wired_point_no_groups = wired.groupby(['point_no'])
print("一共包含  ", len(wired_point_no_groups),"  wired_point_no_groups")
```

```python
# 每组存入csv
i=0
for point_no,point in wired_point_no_groups:
    point = point.sort_values(by=['ts']).loc[:,['ts','acceleration_peak_x','acceleration_rms_x','speed_peak_x','speed_rms_x']]
    point.drop_duplicates(subset=['ts'], inplace=True)
    point.rename(columns={'ts':'timestamp'}, inplace=True)
    point.to_csv('../data_csv/wired_data/wired_point_no_'+str(point_no)+'.csv',index=False)
    print("保存point_no",point_no,"的数据到 Wired_point_no_"+str(point_no)+".csv")
    print(i+1," : 总数据量 ",len(point)," 条")
    print("------------------------------------------------------------")
    print("")
    i=i+1
```
